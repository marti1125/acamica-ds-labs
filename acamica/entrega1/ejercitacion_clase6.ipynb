### Ej 1
- cargar dataset de properati e importar librerias
- hacer un grafico de lineas con la evolucion de los precios segun las fechas .

```python
# Importar bibliotecas en esta celda
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset/datos_properati.csv', parse_dates=["created_on"])
df
```

```python
sns.lineplot(x='created_on', y='price_aprox_usd', data=df) 
```

### Ejercicio 2

- hacer un grafico de caja con los precios en usd por **tipo de propiedad** para el año 2018.

```python
#df.created_on.dt.year
sns.boxplot(x="price_aprox_usd", y="property_type", data=df[df.created_on.dt.year==2018])
```

### Ejercicio 3
- encontrar la cantidad de propiedades de cada barrio, que tienen un precio mayor a 650.000 o mas de 3 habitaciones
- hacer un grafico que lo demuestre

```python
bar = df[(df.price_aprox_usd > 650.000) | (df.rooms > 3)]
bar["property_type"].value_counts()
```

```python
sns.barplot(x="property_type", y="price_aprox_usd", palette="deep", data=bar)
```
